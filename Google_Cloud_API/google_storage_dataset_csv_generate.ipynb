```
import pandas as pd
import numpy as np
import os
```

```
gs_root_path = 'gs://automl-vision-285803-vcm'
```

```
is_root_dir = True
results = []

for full_path,_,images in os.walk('/content/gdrive/My Drive/Colab Notebooks/boat_dataset/'):
  if is_root_dir:
    is_root_dir = False
    continue

  label = ' '.join((os.path.basename(full_path)).split('_')[:2])

  for image in images:
    gs_full_path = gs_root_path + '/' + os.path.basename(full_path) + '/' + image
    results.append([gs_full_path,label])

df = pd.DataFrame(results)
```

```
df.to_csv('/content/gdrive/My Drive/Colab Notebooks/boat_dataset/boat_info.csv',index=False,header=False)
```
